```python
num_bins = 10
plt.figure(figsize=(10,6))
n, bins, patches = plt.hist(polarity, num_bins, facecolor='blue', alpha=0.5)
plt.xlabel('Polarity[1 for positive and -1 for negative]')
plt.ylabel('Count')
plt.title('Histogram of youtube comments analysis')
plt.show();
```
